```python
x=pd.DataFrame(d.iloc[:,2:22].values)
y=d.iloc[:,22].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler  
```

```python
sc = StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#Build the ANN model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
#input layer,2 hidden layer,output layer
classifier.add(Dense(19,activation='relu'))
classifier.add(Dense(15,activation='relu'))
classifier.add(Dense(29,activation='relu'))
classifier.add(Dense(3,activation='linear'))
```

```python
classifier.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
#Train the model
classifier.fit(x_train,y_train,batch_size=10,epochs=25)
```

```python
#Test the model
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
```
